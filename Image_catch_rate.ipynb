````python
import random
# 方案2: 在特定城市随机选择坐标
def generate_random_coordinates(city_center_lat, city_center_lon, radius=0.05):
    """在城市中心附近生成随机坐标"""
    lat = city_center_lat + random.uniform(-radius, radius)
    lon = city_center_lon + random.uniform(-radius, radius)
    return f"{lat},{lon}"


# 纽约市中心示例
ny_locations = [generate_random_coordinates(40.7589, -73.9851) for _ in range(100)]

# 使用你选择的方案
locations = famous_locations  # 或 ny_locations
````

````python
def get_pano_id(location, api_key):
    metadata_url = f"https://maps.googleapis.com/maps/api/streetview/metadata?location={location}&key={api_key}"
    response = requests.get(metadata_url)
    data = response.json()

    if data.get('status') == 'OK':
        return data.get('pano_id')
    return None
````

Create folder and fetch the image

````python
for i in range(num_buildings):
    folder_name = f"data/{i}"
    os.makedirs(folder_name, exist_ok=True)

    location = locations[i % len(locations)]  # 循环使用位置列表
    pano_id = get_pano_id(location, key)

    if pano_id:
        for j, h in enumerate(headings, 1):
            url = f"https://maps.googleapis.com/maps/api/streetview?size=640x640&pano={pano_id}&heading={h}&pitch=0&key={key}"
            img_data = requests.get(url).content
            with open(f"{folder_name}/frame_{j}.jpg", "wb") as f:
                f.write(img_data)

        print(f"✓ 完成文件夹 {i}")
        sleep(0.5)
    else:
        print(f"✗ 跳过文件夹 {i} - 无街景数据")

print("完成！")
````
