````python
import random
# 方案2: 在特定城市随机选择坐标
def generate_random_coordinates(city_center_lat, city_center_lon, radius=0.05):
    """在城市中心附近生成随机坐标"""
    lat = city_center_lat + random.uniform(-radius, radius)
    lon = city_center_lon + random.uniform(-radius, radius)
    return f"{lat},{lon}"


# 纽约市中心示例
ny_locations = [generate_random_coordinates(40.7589, -73.9851) for _ in range(100)]

# 使用你选择的方案
locations = famous_locations  # 或 ny_locations
````

````python
def get_pano_id(location, api_key):
    metadata_url = f"https://maps.googleapis.com/maps/api/streetview/metadata?location={location}&key={api_key}"
    response = requests.get(metadata_url)
    data = response.json()

    if data.get('status') == 'OK':
        return data.get('pano_id')
    return None
````

````python
import PIL.Image
import os
import json
import time
import google.generativeai as genai

# Define base_dir and output_path here, as they are used in this cell
base_dir = "/content/data"
output_path = "/content/scores_teacher.jsonl"

def rate_building(building_id):
    folder_path = os.path.join(base_dir, building_id)

    if not os.path.exists(folder_path):
        print(f"Folder not found: {folder_path}")
        return None

    frames = sorted(
        os.path.join(folder_path, f)
        for f in os.listdir(folder_path)
        if f.lower().endswith(".jpg")
    )

    if not frames:
        print(f"No JPG images found in {folder_path}")
        return None

    # message content
    rating_definition = {
        "rating_definition": {
            "description": (
                "Evaluate the aesthetic quality of the building facade. "
                "Output only: {'score': X.X, 'reason': '...'}"
            ),
            "criteria": [
                {"name": "composition_and_proportion", "weight": 0.25},
                {"name": "material_and_details", "weight": 0.2},
                {"name": "color_harmony", "weight": 0.2},
                {"name": "environmental_harmony", "weight": 0.2},
                {"name": "overall_comfort_and_pressure", "weight": 0.15},
            ],
        }
    }

    # Prepare content for Gemini API
    gemini_content_parts = [
        json.dumps(rating_definition, ensure_ascii=False)
    ]
    for img_path in frames:
        gemini_content_parts.append(PIL.Image.open(img_path))

    # Initialize Gemini model
    model = genai.GenerativeModel('gemini-2.0-flash-lite')

    # retry up to 3 times if API fails
    for attempt in range(3):
        try:
            response = model.generate_content(
                gemini_content_parts,
                generation_config={"temperature": 0.0}
            )
            # Access and clean the text
            text = response.text.strip()
            # Remove markdown code blocks if present
            text = text.replace('```json', '').replace('```', '').strip()
            # Parse JSON
            result = json.loads(text)
            return result
        except Exception as e:
            print(f"Error on {building_id} (attempt {attempt+1}): {e}")
            if attempt < 2:
                print("Retrying in 5s...")
                time.sleep(5)
            else:
                return None


# load existing results to resume if needed
existing = {}
if os.path.exists(output_path):
    with open(output_path, "r", encoding="utf-8") as f:
        for line in f:
            item = json.loads(line)
            existing[item["building_id"]] = item["score_teacher"]

building_ids = sorted(
    d for d in os.listdir(base_dir)
    if os.path.isdir(os.path.join(base_dir, d))
)

with open(output_path, "a", encoding="utf-8") as f:
    for building_id in building_ids:
        if building_id in existing:
            print(f"Skipping {building_id} (already done)")
            continue

        result = rate_building(building_id)
        if result is None:
            continue

        record = {"building_id": building_id, "score_teacher": result}
        f.write(json.dumps(record, ensure_ascii=False) + "\n")
        f.flush()
        print(f"{building_id}: {result}")

print("Processing complete!")
````

````python
import google.generativeai as genai

model_name = 'gemini-2.0-flash-lite'
model = None
for m in genai.list_models():
    if m.name == f'models/{model_name}':
        model = m
        break

if model:
    print(f"Details for {model_name}:\n")
    print(f"  Name: {model.name}")
    print(f"  Display Name: {model.display_name}")
    print(f"  Supported Generation Methods: {model.supported_generation_methods}")
    print(f"  Input Token Limit: {model.input_token_limit}")

    # Check for image input support more robustly
    has_image_input = False
    if hasattr(model, 'input_token_limit') and model.input_token_limit:
        if isinstance(model.input_token_limit, list): # Check if it's a list of limits
            for limit_obj in model.input_token_limit:
                if hasattr(limit_obj, 'name') and 'image' in limit_obj.name.lower():
                    has_image_input = True
                    break
        elif isinstance(model.input_token_limit, int): # If it's just an int, it's typically text only
             has_image_input = False # An int usually means total token limit, not specifying image types
        else: # Try to inspect as a generic object
             if hasattr(model.input_token_limit, 'name') and 'image' in model.input_token_limit.name.lower():
                 has_image_input = True

    if has_image_input:
        print(f"  Supports Image Input: Yes")
    else:
        print(f"  Supports Image Input: No (based on input_token_limit inspection)")
else:
    print(f"Model {model_name} not found.")
````
